```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

sns.set()
sns.set_style("whitegrid")

df = pd.read_csv('data_bike/citybike_Feb-Jun.csv')
```

```python
colors = np.where(df['citibike']<0, 'negative', 'positive')
palette=sns.color_palette(['purple','green'])
sns.swarmplot(x="month", y="citibike", data=df, s=2.5, hue=colors, palette=palette)
plt.xlabel('NEW YORK CITY')
plt.show()
```

```python
df_BRONX = df.drop(df[df['borough'] != 'Brooklyn'].index, inplace = False)
colors = np.where(df_BRONX['citibike']<0, 'negative', 'positive')
palette=sns.color_palette(['purple','green'])
sns.swarmplot(x="month", y="citibike", data=df_BRONX, s=2.5, hue=colors, palette=palette)
plt.xlabel('BROOKLYN')
plt.show()
```

```python
df_BROOKLYN = df.drop(df[df['borough'] != 'Manhattan'].index, inplace = False)
colors = np.where(df_BROOKLYN['citibike']<0, 'negative', 'positive')
palette=sns.color_palette(['green','purple'])
sns.swarmplot(x="month", y="citibike", data=df_BROOKLYN, s=2.5,hue=colors, palette=palette)
plt.xlabel('MANHATTAN')
plt.show()
```
