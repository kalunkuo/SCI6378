```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

sns.set()
sns.set_style("whitegrid")

df = pd.read_csv('data_sets/mobility_summary_Feb-Jun.csv')
```

```python
colors = np.where(df['walking']<0, 'negative', 'positive')
palette=sns.color_palette(['pink','lightgreen'])
sns.swarmplot(x="month", y="walking", data=df, s=2.5, hue=colors, palette=palette)
plt.xlabel('NEW YORK CITY')
plt.show()
```
