```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_style("whitegrid")

%precision 3
pd.set_option('precision', 3)

# Enable Jupyter Notebook's intellisense
%config IPCompleter.greedy=True
```

```python
file_path='/Users/karenkuo/file/data/mobility/apple mobility trend_ny_2.csv'
mob2 = pd.read_csv(file_path, parse_dates=['Date'], index_col=['Date'])
mob2.head()
```

```python
fig, ax = plt.subplots(figsize=(30, 10))
ax.plot(mob2.index.values, mob2['driving'], color='orange',alpha=0.7)
ax.plot(mob2.index.values, mob2['walking'], color='royalblue',alpha=0.7)
ax.plot(mob2.index.values, mob2['transit'], color='Grey',alpha=0.7)

ax.set(xlabel="Date", ylabel="driving / walking / transit", title="Mobility trends")
plt.savefig('appleMobilityTrend.pdf', 
           dpi=300,bbox_inches="tight")
```

```python
plt.subplots(figsize=(30,10))

sns.scatterplot(x='Date', y='driving', data=mob2)
g=sns.scatterplot(x='Date', y='transit', data=mob2)
g=sns.scatterplot(x='Date', y='walking', data=mob2)

plt.xticks(rotation=60)
plt.ylabel("Mobility")

#sns.replot(x="Month", y="driving", dashes=False, markers=True, kind="line", data=mob2);
#sns.relplot(x="Date", y="transit", dashes=False, markers=True, kind="line", data=mob2);
#sns.relplot(x="Date", y="walking", dashes=False, markers=True, kind="line", data=mob2);
```
